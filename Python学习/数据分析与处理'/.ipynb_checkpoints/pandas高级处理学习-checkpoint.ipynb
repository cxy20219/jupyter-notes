### 1.1 缺失值处理  (NaN)
* 数据量大时可以删除掉有缺失值的样本
* 替换缺失值

```python
import pandas as pd
```

```python
movie = pd.read_csv("./IMDB/IMDB-Movie-Data.csv")
```

```python
# 1）判断是否存在NaN类型的缺失值
movie.head()
```

```python
import numpy as np
```

```python
np.any(pd.isnull(movie)) # 返回True，说明数据中存在缺失值 # 有True时返回True
```

```python
np.all(pd.notnull(movie)) # 返回False，说明数据中存在缺失值 # 有False返回False
```

```python
pd.isnull(movie).any()   # 有True时返回True
```

```python
pd.notnull(movie).all()  # 有False返回False
```

```python
# 2）缺失值处理
# 方法1：删除含有缺失值的样本
data1 = movie.dropna()
```

```python
pd.notnull(movie).all()
```

```python
pd.notnull(data1).all()
```

```python
# 方法2：替换
movie.head()
```

```python
# 含有缺失值的字段
# Revenue (Millions)    
# Metascore

movie["Revenue (Millions)"].fillna(movie["Revenue (Millions)"].mean(), inplace=True) # 为True 直接在原数据修改 默认False
movie["Metascore"].fillna(movie["Metascore"].mean(), inplace=True)
```

```python
movie.head()
```

```python
pd.notnull(movie).all() # 缺失值已经处理完毕，不存在缺失值
```

### 1.2 不是缺失值nan，有默认标记的

```python
# 读取数据
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
name = ["Sample code number", "Clump Thickness", "Uniformity of Cell Size", "Uniformity of Cell Shape", "Marginal Adhesion", "Single Epithelial Cell Size", "Bare Nuclei", "Bland Chromatin", "Normal Nucleoli", "Mitoses", "Class"]


data = pd.read_csv(path, names=name)
```

```python
data.head()
```

```python
# 1）替换
data_new = data.replace(to_replace="?", value=np.nan)
```

```python
data_new.head()
```

```python
# 2）删除缺失值
data_new.dropna(inplace=True)
```

```python
data_new.isnull().any() # 全部返回False说明不存在缺失值了
```

```python
type(np.nan)
```

### 2.1 如何实现数据的离散化

```python
# 1）准备数据
data = pd.Series([165,174,160,180,159,163,192,184], index=['No1:165', 'No2:174','No3:160', 'No4:180', 'No5:159', 'No6:163', 'No7:192', 'No8:184']) 
```

```python
data
```

```python
# 2）分组
# 自动分组
sr = pd.qcut(data, 3)
```

```python
type(sr)
```

```python
sr
```

```python
sr.value_counts()
```

```python
# 3）转换成one-hot编码
pd.get_dummies(sr, prefix="height")
```

```python
# 自定义分组
bins = [150, 165, 180, 195]
sr = pd.cut(data, bins)
```

```python
sr
```

```python
sr.value_counts()
```

```python
# get_dummies
pd.get_dummies(sr, prefix="身高")
```

### 2.2 案例：股票的涨跌幅离散化

```python
# 1）读取数据
stock = pd.read_csv("./stock_day/stock_day.csv")
```

```python
stock.head()
```

```python
p_change = stock["p_change"]
```

```python
p_change.head()
```

```python
# 2）分组
sr = pd.qcut(p_change, 10)
```

```python
sr.value_counts()
```

```python
# 3）离散化
pd.get_dummies(sr, prefix="涨跌幅")
```

```python
# 自定义分组
bins = [-100, -7, -5, -3, 0, 3, 5, 7, 100]
sr = pd.cut(p_change, bins)
```

```python
sr.value_counts()
```

```python
# one-hot
pd.get_dummies(sr, prefix="rise").head()
```

```python
stock_change = pd.get_dummies(sr, prefix="rise")
```

```python
stock_change
```

### 3.2 pd.merge合并
按索引拼接

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how="inner", on=["key1", "key2"]) # 保留共有的键
```

```python
pd.merge(left, right, how="left", on=["key1", "key2"])  
```

```python
pd.merge(left, right, how="outer", on=["key1", "key2"])  
```

### 4.1 使用crosstab(交叉表)实现
探索两个变量之间的关系

```python
stock.head()
```

```python
# 星期数据以及涨跌幅是好是坏数据
# pd.crosstab(星期数据列, 涨跌幅数据列)
# 准备星期数据列
stock.index
```

```python
# pandas日期类型
date = pd.to_datetime(stock.index)
```

```python
date
```

```python
stock["week"] = date.weekday
```

```python
date.weekday
```

```python
stock.head()
```

```python
# 准备涨跌幅数据列
stock["pona"] = np.where(stock["p_change"] > 0, 1, 0)
```

```python
stock.head()
```

```python
# 交叉表
data = pd.crosstab(stock["week"], stock["pona"])
```

```python
data
```

```python
data.sum(axis=1)
```

```python
data.div(data.sum(axis=1), axis=0)
```

```python
data.div(data.sum(axis=1), axis=0).plot(kind="bar", stacked=True)  # 同一x轴绘制不同的y值
```

```python
# 透视表操作
stock.pivot_table(["pona"], index=["week"])
```

### 5.1 分组与聚合

```python
col =pd.DataFrame({'color': ['white','red','green','red','green'], 'object': ['pen','pencil','pencil','ashtray','pen'],'price1':[5.56,4.20,1.30,0.56,2.75],'price2':[4.75,4.12,1.60,0.75,3.15]})
```

```python
col
```

```python
# 分组后没有直接显示，需要聚合表现出来
# 进行分组，对颜色分组，price1进行聚合
# 用dataframe的方法进行分组
col.groupby(by="color")["price1"].max()
```

```python
col["price1"].groupby(col["color"]).max()
```

### 5.2 星巴克零售店铺数据案例

```python
# 1、准备数据
starbucks = pd.read_csv("directory.csv")
```

```python
starbucks.head()
```

```python
# 按照国家分组，求出每个国家的星巴克零售店数量
starbucks.groupby("Country").count()["Brand"].sort_values(ascending=False)[:10].plot(kind="bar", figsize=(20, 8), fontsize=40)
```

```python
# 假设我们加入省市一起进行分组
starbucks.groupby(by=["Country", "State/Province"]).count().head()
```
