````python
from torch.utils.tensorboard import SummaryWriter
````

### 启用tensorboard

* 启动：
    ```bash
    tensorboard --logdir=./tensorboard_logs
    ```
* 更改启动端口：
    ```bash
    tensorboard --logdir=./tensorboard_logs --port=10006
    ```

### 标量数据
**实例化对象**

````python
writer = SummaryWriter("./tensorboard_logs")
````

**设置坐标轴、标签**

````python
for i in range(100):
    writer.add_scalar("y=x",i,i)
````

````python
for i in range(100):
    writer.add_scalar("y=2x",i**2,i)
````

### 图片数据

````python
img_path = "../../../datas/hymenoptera_data/train/ants/0013035.jpg"
from PIL import Image

img = Image.open(img_path)
img
````

````python
print(type(img))
# opencv读取的图像是numpy形
````

**转换图片数据类型**

````python
import numpy as np
img = np.array(img)
img
````

````python
img.shape              # HWC 高度H 宽度W 通道C
````

````python
writer.add_image("ant01",img,1,dataformats="HWC")  # 默认为CHW # 第几步
````

````python
writer.close()
````
