```python
image_path = "Test_Images/horse.png"
```

```python
from PIL import Image
```

```python
image = Image.open(image_path)
image
```

```python
print(image)
```

**将输入图片转换模式**

```python
# png图片是四通道转换为RBG
if(image.mode!='RGB'):        
    image = image.convert("RGB")
```

```python
import torchvision
import torch
```

```python
transform = torchvision.transforms.Compose([
    torchvision.transforms.Resize((32,32)),
    torchvision.transforms.ToTensor()
])
```

```python
T_image = transform(image)
T_image.shape
```

```python
T_image = torch.reshape(T_image,(1,3,32,32))
T_image.shape
```

**加载网络模型**

```python
class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.model = torch.nn.Sequential(
            torch.nn.Conv2d(in_channels=3,out_channels=32,kernel_size=5,stride=1,padding=2),
            torch.nn.MaxPool2d(kernel_size=2),
            torch.nn.Conv2d(32,32,5,1,2),
            torch.nn.MaxPool2d(2),
            torch.nn.Conv2d(32,64,5,1,2),
            torch.nn.MaxPool2d(2),
            torch.nn.Flatten(),
            torch.nn.Linear(64*4*4,64),
            torch.nn.Linear(64,10)
        )
        
    def forward(self,x):
        x = self.model(x)
        return x
```

```python
model = torch.load("models/model_10.pth")
```

```python
model
```

```python
output = model(T_image)
output
```

```python
text_list = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']
```

```python
text_list[output.argmax(1)]
```
