```python
from torch.utils.data import DataLoader
```

```python
import torchvision
```

```python
train_set = torchvision.datasets.CIFAR10(
    root = "../../../datas",
    train = True,
    transform = torchvision.transforms.ToTensor(),
    download = True
)
```

```python
test_set = torchvision.datasets.CIFAR10(
    root = "../../../datas",
    train = False,
    download = True,
    transform = torchvision.transforms.ToTensor()
)
```

```python
test_loader = DataLoader(
    dataset = test_set,
    batch_size = 4,   # 每次从中取四个数据集打包
    shuffle = True,   # 不按顺序取
    num_workers = 0,  # 设置进程数 0:核心进程
    drop_last = False
)
```

```python
img,target = test_set[0]
img.shape
```

```python
for data in test_loader:
    imgs,targets = data
    print(imgs.shape,end=" ")
    print(targets)
```

```python
from torch.utils.tensorboard import SummaryWriter
writer = SummaryWriter("./tensorboard_dataloader")
step = 1
for data in test_loader:
    imgs,targets = data
    writer.add_images("test_data",imgs,step)
    step+=1
```
