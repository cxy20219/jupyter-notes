识别图像主体的预训练网络
======================== 

### 1. 获取预训练好的网络

```python
from torchvision import models
```

查看模型

* 大写字母开头名称对应流行模型的python类
* 小写字母开头指预定义的层和单元数实例化函数

*resnet101 表示用来实例化具有101层的卷积神经网络* 

```python
dir(models)
```

```python
# 实例化对象
alexnet = models.AlexNet()
```

```python
restnet = models.resnet101(pretrained=True)
```

```python
restnet
```

### 2. 处理输入图像

```python
from torchvision import transforms
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),     # 将图像缩放到256 X 256个像素
    transforms.CenterCrop(224), # 围绕中心将图像剪裁为224 X 224个像素
    transforms.ToTensor(),      # 转换为张量
    transforms.Normalize(
            mean=[0.485,0.456,0.406],   # 定义均值
            std=[0.229,0.224,0.225]     # 定义标准差
    )  
])
```

```python
from PIL import Image
img = Image.open("data/cat.jpg")
img
```

```python
# 通过预处理管道处理传递图像
img_t = preprocess(img)
img_t
```

```python
import torch
batch_t = torch.unsqueeze(img_t,0)
batch_t
```

将网络置于eval模式

```python
restnet.eval()
```

```python
out = restnet(batch_t)
out
```

```python
with open("data/imagenet_classes.txt") as f:
    labels = [line.strip() for line in f.readlines()]
labels
```

```python
_,index = torch.max(out,1)
index
```

```python
index[0]
```

将输出归一化到(0,1)

```python
precentage = torch.nn.functional.softmax(out,dim=1)[0]*100 
labels[index[0]],precentage[index[0]].item()
```

```python
_,indices = torch.sort(out,descending = True)
[(labels[idx],precentage[idx].item())for idx in indices[0][:5]]
```
