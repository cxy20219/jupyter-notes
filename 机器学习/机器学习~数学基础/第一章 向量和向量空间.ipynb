3. 运算速度示例

```python
import random,time
# 创建列表
lst = [random.randint(1,100) for i in range(1000000)]

start = time.time()
lst = [i*i for i in lst]  # 列表形式计算平方
end = time.time()
print(f"用时：{end - start}s")

vlst = np.array(lst)      # 列表化为数组
start2 = time.time()
vlst2 = vlst*vlst         # 用数组计算平方
end2 = time.time();
print(f"用时：{end2 - start2}s")

print(f"数组用时快{round((end - start)/(end2 - start2),3)}倍")
```

## 4. 距离和角度

### 4.1 距离

* 欧几里得距离



```python
import numpy as np
vec1 = np.array([1,2])
vec2 = np.array([9,8])
dist = np.linalg.norm(vec1-vec2)
dist
```

* 曼哈顿距离


```python
from scipy.spatial.distance import cityblock
a = np.array([1,5])
b = np.array([6,2])
md = cityblock(a,b)
md
```

* 切比雪夫距离


```python
from scipy.spatial.distance import chebyshev
c = np.array([1,5])
d = np.array([6,2])
cf = chebyshev(c,d)
cf
```

* 闵可夫斯基距离

1. p=1,曼哈顿距离
2. p=2,欧几里得距离
3. p->∞，切比雪夫距离
